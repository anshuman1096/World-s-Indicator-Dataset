# Exploring GDP per capita vs CO2 emissions

```python
%matplotlib inline
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./Indicators.csv')
data.shape
```

```python
data.head(10)
```

```python
countries=data['CountryName'].unique().tolist()
len(countries)
```

```python
countries=data['CountryCode'].unique().tolist()
len(countries)
```

Here we'll use <b>Indicator = 'CO2 emissions'</b> and <b>country = 'USA'</b>

```python
hist_indicator = 'CO2 emissions \(metric'
hist_country = 'USA'

mask1 = data['IndicatorName'].str.contains(hist_indicator)
mask2 = data['CountryCode'].str.contains(hist_country)

stage = data[mask1 & mask2]
```

```python
stage
```

```python
year  = stage['Year'].values
co2 = stage['Value'].values

plt.bar(year,co2)
plt.show()
```

```python
plt.plot(stage['Year'].values,stage['Value'].values)

plt.xlabel('Year')
plt.ylabel(stage['IndicatorName'].iloc[0])

plt.title('CO2 Emissions in USA')

plt.axis([1956,2011,0,24])

plt.show()
```

```python
hist_data = stage['Value'].values
```

```python
hist_data.shape
```

```python
plt.hist(hist_data,10,density=False,facecolor='green')
plt.xlabel(stage['IndicatorName'].iloc[0])
plt.ylabel('# of Years')
plt.title('Histogram Example')

plt.grid(True)
plt.show()
```

```python
hist_ind = 'CO2 emissions \(metric'
hist_year = '2011'

mask1 = data['IndicatorName'].str.contains(hist_ind)
mask2 = data['Year'].isin([hist_year])

co2_2011 = data[mask1 & mask2]
```

```python
fig,ax = plt.subplots()

ax.annotate("USA",
            xy=(18,5), xycoords='data',
            xytext=(18,30), textcoords='data',
            arrowprops=dict(arrowstyle=" -> ",
                            connectionstyle="arc3")
)


plt.hist(co2_2011['Value'].values,10,density=False,facecolor='green')
plt.xlabel(stage['IndicatorName'].iloc[0])
plt.ylabel('# of countries')
plt.title('Histogram of CO2 emissions per capita')
plt.grid(True)
plt.show()
```

```python
hist_ind = 'GDP per capita \(constant 2005'
hist_country = 'USA'

mask1 = data['IndicatorName'].str.contains(hist_ind)
mask2 = data['CountryCode'].str.contains(hist_country)

gdp = data[mask1 & mask2]
gdp.shape
```

```python
plt.plot(gdp['Year'].values, gdp['Value'].values)

plt.xlabel('Year')
plt.ylabel(gdp['IndicatorName'].iloc[0])

plt.title('GDP per capita of USA')

plt.show()
```

```python
print("gdp min and max",gdp['Year'].min(),gdp['Year'].max())
print("co2 min and max",stage['Year'].min(),stage['Year'].max())
```

```python
gdp_trunc = gdp[gdp['Year'] < 2012] 
print(len(gdp_trunc))
print(len(stage))
```

```python
fig,axis = plt.subplots()

axis.yaxis.grid(True)
axis.set_title('CO2 emissions vs GDP \(per capita\)', fontsize=10)
axis.set_xlabel(gdp_trunc['IndicatorName'].iloc[0], fontsize=10)
axis.set_ylabel(stage['IndicatorName'].iloc[0], fontsize=10)

X=gdp_trunc['Value'].values
Y=stage['Value'].values

axis.scatter(X,Y)
plt.show()
```
